```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
website = "https://en.wikipedia.org/wiki/List_of_power_stations_in_Singapore"
source = requests.get(website).text
data = BeautifulSoup(source, 'lxml')
```

```python
column_names = ['Power Station','Capacity','Reference']
pss = pd.DataFrame(columns = column_names)

content = data.find('div', class_='mw-parser-output')
table = content.table.tbody
ps = 0
cap = 0
ref = 0

for tr in table.find_all('tr'):
    i = 0
    for td in tr.find_all('td'):
        if i == 0:
            ps = td.text
            i = i + 1
        elif i == 1:
            cap = td.text
            i = i + 1
        elif i == 2: 
            ref = td.text.strip('\n').replace('{}','')
    pss = pss.append({'Power Station': ps,'Capacity': cap,'Reference': ref},ignore_index=True)

pss = pss[pss.Capacity!='Not assigned']
pss = pss[pss.Capacity!= 0]
pss.reset_index(drop = True, inplace = True)
i = 0
for i in range(0,pss.shape[0]):
    if pss.iloc[i][2] == 'Not assigned':
        pss.iloc[i][2] = pss.iloc[i][1]
        i = i+1
                                 
df = pss.groupby(['Power Station','Capacity'])['Reference'].apply(''.join).reset_index()
df.head(50)
```
